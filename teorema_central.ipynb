```python
import numpy.random as npr
import numpy as np
import pandas as pd
```

```python
npr.seed(1)
muestra_binomial= npr.binomial(1,0.6,1000)
muestra_binomial
```

```python
df=pd.DataFrame()

for i in range(1,100):
    muestras_binomiales=npr.binomial(1000,0.6,50)
    col = f"Muestras_{i}"
    df[col]=muestras_binomiales

df

df_media=pd.DataFrame(df.mean(),columns=["Muestra Medias"])

df_media
```
